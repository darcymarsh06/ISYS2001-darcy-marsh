## Part 2: Your First AI Prompt

Ask an AI assistant (Gemini in Colab, ChatGPT, or GitHub Copilot) to write a program that:
- Asks the user for their name
- Asks for their monthly income
- Prints a personalised welcome message

Write your prompt as a comment, then paste in whatever code the AI gives you back.


```python
# Write your AI prompt here as a comment:
"""
My AI Prompt:
[write a program that
- asks the user for their name
- asks for their monthly income
- prints a personalised welcome message]
"""

# Paste the AI-generated code below:
def main():
    # Ask the user for their name
    name = input("Enter your name: ").strip()

    # Ask for their monthly income with basic error handling
    while True:
        try:
            income = float(
                input("Enter your monthly income: ").replace("$", "").strip()
            )
            if income < 0:
                print("Income cannot be negative. Please try again.")
                continue
            break
        except ValueError:
            print(
                "Invalid input. Please enter a valid number (e.g., 4500 or 4500.50)."
            )

    # Print a personalized welcome message
    print("\n" + "=" * 45)
    print(f"Welcome, {name}!")
    print(
        f"We've recorded your monthly income as ${income:,.2f}."
    )
    print("=" * 45)


if __name__ == "__main__":
    main()
```

## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income
- Takes up to 3 expenses
- Calculates what's left over
- Warns the user if the budget goes negative

```python
# Write your detailed prompt here:
"""
My Budget Tracker Prompt:
[ you write a program that:

Takes the user's monthly income
Takes up to 3 expenses
Calculates what's left over
Warns the user if the budget goes negative]
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
def get_valid_amount(prompt):
    """Helper function to safely get positive numbers from the user."""
    while True:
        try:
            val = float(input(prompt).replace("$", "").strip())
            if val < 0:
                print("Amount cannot be negative. Please try again.")
                continue
            return val
        except ValueError:
            print("Invalid input. Please enter a valid number (e.g., 1500 or 1500.50).")


def main():
    print("=== Monthly Budget Calculator ===\n")

    # 1. Get monthly income
    income = get_valid_amount("Enter your monthly income: ")

    # 2. Collect up to 3 expenses
    print("\nEnter up to 3 monthly expenses (press Enter with no value to stop early):")
    expenses = []

    for i in range(1, 4):
        raw_input = input(f"Expense {i} amount (or press Enter to skip): ").replace("$", "").strip()

        # Allow user to stop early if they have fewer than 3 expenses
        if raw_input == "":
            print("Finished adding expenses.")
            break

        try:
            val = float(raw_input)
            if val < 0:
                print("Expense cannot be negative. Skipping this entry.")
                continue
            expenses.append(val)
        except ValueError:
            print("Invalid number entered. Skipping this entry.")

    # 3. Calculate total expenses and remaining balance
    total_expenses = sum(expenses)
    remaining_balance = income - total_expenses

    # 4. Print summary & warning if negative
    print("\n" + "=" * 40)
    print("           BUDGET SUMMARY           ")
    print("=" * 40)
    print(f"Monthly Income:    ${income:,.2f}")
    print(f"Total Expenses:    ${total_expenses:,.2f}")
    print(f"Remaining Balance: ${remaining_balance:,.2f}")
    print("-" * 40)

    if remaining_balance < 0:
        print("⚠️ WARNING: Your budget is in the negative!")
        print(f"You are over budget by ${abs(remaining_balance):,.2f}.")
    elif remaining_balance == 0:
        print("ℹ️ You have broken perfectly even this month.")
        print("Consider allocating a small amount toward savings next time.")
    else:
        print("✅ Great job! You have a positive budget surplus.")

    print("=" * 40)


if __name__ == "__main__":
    main()
```

Pick ONE feature to add with AI's help:

```python
# ENHANCED BUDGET TRACKER
def get_valid_amount(prompt):
    """Helper function to safely get positive numbers from the user."""
    while True:
        try:
            val = float(input(prompt).replace("$", "").strip())
            if val < 0:
                print("Amount cannot be negative. Please try again.")
                continue
            return val
        except ValueError:
            print(
                "Invalid input. Please enter a valid number (e.g., 1500 or 1500.50)."
            )


def main():
    print("=== Monthly Budget Calculator ===\n")

    # 1. Get monthly income
    income = get_valid_amount("Enter your monthly income: ")

    # 2. Collect up to 3 expenses
    print(
        "\nEnter up to 3 monthly expenses (press Enter with no value to stop early):"
    )
    expenses = []

    for i in range(1, 4):
        raw_input = (
            input(f"Expense {i} amount (or press Enter to skip): ")
            .replace("$", "")
            .strip()
        )

        # Allow user to stop early if they have fewer than 3 expenses
        if raw_input == "":
            print("Finished adding expenses.")
            break

        try:
            val = float(raw_input)
            if val < 0:
                print("Expense cannot be negative. Skipping this entry.")
                continue
            expenses.append(val)
        except ValueError:
            print("Invalid number entered. Skipping this entry.")

    # 3. Calculate metrics
    total_expenses = sum(expenses)
    remaining_balance = income - total_expenses

    # Calculate percentage spent (protecting against division by zero if income is 0)
    percent_spent = (
        (total_expenses / income) * 100 if income > 0 else 0.0
    )

    # 4. Print summary & warnings
    print("\n" + "=" * 40)
    print("           BUDGET SUMMARY           ")
    print("=" * 40)
    print(f"Monthly Income:    ${income:,.2f}")
    print(f"Total Expenses:    ${total_expenses:,.2f}")
    print(f"Remaining Balance: ${remaining_balance:,.2f}")
    print(f"Income Spent:      {percent_spent:.1f}%")
    print("-" * 40)

    if remaining_balance < 0:
        print("⚠️ WARNING: Your budget is in the negative!")
        print(
            f"You are spending {percent_spent:.1f}% of your income (over budget by ${abs(remaining_balance):,.2f})."
        )
    elif remaining_balance == 0:
        print("ℹ️ You have spent 100% of your income and broken perfectly even.")
        print("Consider allocating a small amount toward savings next time.")
    else:
        print(
            f"✅ Great job! You saved {100 - percent_spent:.1f}% of your income."
        )

    print("=" * 40)


if __name__ == "__main__":
    main()
```
